```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
```

Load CO2 emission data by country

```python
country_co2 = pd.read_csv('CO2EmissionsByCountry.csv', skiprows=4)
```

```python
dates_corrected = []

for date in country_co2.DATE:
    dates_corrected.append(datetime.date(year=int(date[6:]), month=int(date[3:5]), day=int(date[:2])))

country_co2['Date_'] = dates_corrected
```

```python
fig, ax = plt.subplots(figsize=(16,12))
ax.plot(country_co2['Date_'], country_co2['United Kingdom'], linewidth=2, color = 'r', alpha=0.7)
#ax.fill_between(country_co2['year'], country_co2['low uncertainty'], country_co2['high uncertainty'],facecolor='r',alpha=0.3)
ax.invert_yaxis()
ax.set_title('United Kingdom daily CO2 Emissions')
ax.set_xlabel('Date')
ax.set_ylabel('CO2 Emissions [MtCO2 per day]')
locator = mdates.AutoDateLocator()
ax.xaxis.set_major_locator(locator)
```

Load global daily CO2 Emissions data

```python
daily_co2 = pd.read_csv('GlobalDailyCO2.csv',skiprows=4)
```

```python
fig, ax = plt.subplots(figsize=(16,12))
ax.plot(daily_co2['year'], daily_co2['value'], linewidth=2, color = 'r', alpha=0.7)
ax.fill_between(daily_co2['year'], daily_co2['low uncertainty'], daily_co2['high uncertainty'],facecolor='r',alpha=0.3)
ax.set_title('Global daily CO2 Emissions')
ax.set_xlabel('Date [Years]')
ax.set_ylabel('CO2 Emissions [MtCO2 per day]')
```
