# People, Society and Climate Change 
Despite all the myriad of words, discussions, research and arguments, the Paris Climate Accord has a simple fundamental aim to reduce atmospheric carbon dioxide to a level that will limit global temperature rise. 

However, the potential solutions are numerous, highly interconnected and far from simple to implement and achieve. Team Coronasaurus has therefore concentrated on just one strand of a potential solution based upon the real-world experience of the 2020 Covid 19 UK lockdown (March 2020 – to date). 

Lockdown, during the Covid 19 epidemic, forced everyone in the population to radically adapt and change their lifestyle, work and social habits. There is not a single household or workplace that the Covid 19 Lockdown did not have an impact upon. Many of these adaptations were forced upon them through government policy (from closing shops, schools and pubs through to instructing certain people to remain at home and not going outside). However, being human, the actions households and people took were based upon their interpretation, readiness to adapt and willingness to comply with these policies. Going into, but especially coming out of Lockdown, people complied with the rules in their  own unique way. 

Team Coronasaurus examined the effects of the lockdown, seen through publicly available data, to examine how reducing and changing lifestyle, living arrangements, social interaction and work patterns (to incorporate more remote working) could contribute towards long term carbon dioxide reduction. 

* How well did people conform, adapt and permanently alter their behaviours to this radical situation ? 

* What lessons can be learnt to assist the implement the radical lifestyle changes needed to address Climate Change ?

* How can society help people to adapt and adjust and even see opportunity in necessary changes ?

**How did domestic problems effect people during Lockdown and how did these issues alter as Lockdown was eased ?**

The Greater London Authority conducted weekly surveys from 20 April 2020 onwards to assess the impact of Covid 19 including how the population was coping. Part of this effort was achieved by surveying civil society organisations (charities, local authorities, NGOs, organisation, help groups, etc) and asking them how many people had contacted them for help which they had then either provided or assisted with (the important criteria was that action was actually taken over the problem which eliminated enquiries then may have not actually needed external assistance to solve).

The figures quoted are expressed as a percentage of people helped greater than during normal times. 

These figures therefore represent the direct social effects of Covid 19 and the Lockdown. 

Data Source: Greater London Authority.  https://data.london.gov.uk/dataset/london-community-response-survey




```python
df['Victim support'] = df['Victim support'].fillna(0)
```

```python
df['Victim support'] = df['Victim support'].astype('int64') 
```

```python
df['Date'] = to_datetime(df['Date'])
```

```python
%matplotlib inline
```

```python
ax=df[['Access to benefits', 	'Access to care', 	'Access to food', 	'Access to health']].plot(grid=True, figsize=(10,5),title='Meeting Basic Needs to Survive')
ax.set_ylabel('% of actions delivered above normal levels')
bx=df[['Mental health', 'Isolation and loneliness','Physical health','Poverty']].plot(grid=True, figsize=(10,5),title='Meeting Individual Needs to Thrive')
bx.set_ylabel('% of actions delivered above normal levels')
cx=df[['Childcare or Parenting', 'Debt issues','Domestic violence','Grief bereavement']].plot(grid=True, figsize=(10,5),title='Thriving as a Household')
cx.set_ylabel('% of actions delivered above normal levels')
dx=df[['Discrimination','Immigration issues','Housing', 'Victim support']].plot(grid=True, figsize=(10,5),title='Thriving in the Community')
dx.set_ylabel('% of actions delivered above normal levels')
ex=df[['Digital connectivity','Employment','Capacity Building']].plot(grid=True, figsize=(10,5),title='Interaction with Society')
ex.set_ylabel('% of actions delivered above normal levels')
```

Characterising the trends over the lockdown period of each type of action there are 2 classifications that can be made:

* Not much change - within 10% to 20 % band with no distinctive trend - possibly classified as more long term and intractable issues such as immigration and housing.
* Consistent downward trend over the lockdown period - this tends to be those "fundamental to life issues" and "fundenental to an individual functioning" such as access to food, benefits, healthcare, etc. These are the base survival elements of Maslow's triangle which individuals and outside agencies will target to solve as they must be achieved for wellbeing. These tend to be problems that must be addressed in a short timescale.

## Happiness, contentment and Anxiety
Over two-thirds of adults consistantly have said that they were very or somewhat worried about the effect that coronavirus (COVID-19) is having on their life.

Through the weeks of lockdown, of the four measures of personal well-being, falling anxiety levels have seen the largest change over the period with happiness increasing over time.

Data Source: Office for National Statistics (ONS).  https://www.ons.gov.uk/peoplepopulationandcommunity/healthandsocialcare/healthandwellbeing/bulletins/coronavirusandthesocialimpactsongreatbritain/12june2020


Notes:
* Question: "Overall, how satisfied are you with your life nowadays?", "Overall, to what extent do you feel that the things you do in your life are worthwhile?", "Overall, how happy did you feel yesterday?", "Overall, how anxious did you feel yesterday?".
* Each of these questions is answered on a scale of 0 to 10, where 0 is “not at all” and 10 is “completely”.

```python
import numpy as np
import matplotlib.pyplot as plt
df1 = read_csv('ONS_happiness.csv')
```

```python
df1.index = df1['Dates']
df1['Dates'] = to_datetime(df1['Dates'])
```

```python
ax=df1[['Life satisfaction', 'Feeling worthwhile', 'Happiness', 'Anxiety']].plot(grid=True, figsize=(10,5),title='ONS Opinions and Lifestyle Survey​ during Lockdown')
ax.set_ylabel('Survey Response Scale. 0 = "Not at All", 10 = "Completely"')
```

Characterising the trends over the lockdown period of each type of action there are 2 classifications that can be made:

* Not much change - these are wellness indicators or happiness, life satisfaction and feelings of worth. It appears that whilst lockdown brought drastic change to lifestyle along with some degree of isolation it has made little difference to peoples self value.
* Anxiety caused by lockdown has shown a consistent downward trend - this would indicate that people will eventually get used to change.

## Conclusions from analysis

* Civil Society and individuals will adapt to change in order to meet their basic needs. 

* People will tend to adjust and eventually thrive (based on their anxiety and happiness measures) during periods of long term change and upheaval. 

* It is hypothosised that this is helped by everyone being in the same situation of having to adapt to change.
